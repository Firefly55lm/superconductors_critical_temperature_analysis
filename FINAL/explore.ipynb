### CHARTS

The following graphs isolate two selected macro-variables and plot the correlation matrix (corrplot) for the respective indexes in the dataset.

This is useful to observe the correlation between the metrics. Although, they have been calculated by the dataset's author because considered useful to
explain the variability of the phenomenon.

```python
columns_to_keep = ['mean_atomic_mass', 'wtd_mean_atomic_mass',
       'gmean_atomic_mass', 'wtd_gmean_atomic_mass', 'entropy_atomic_mass',
       'wtd_entropy_atomic_mass', 'range_atomic_mass', 'wtd_range_atomic_mass',
       'std_atomic_mass', 'wtd_std_atomic_mass']

corr_matrix = np.corrcoef(data[columns_to_keep], rowvar=False)
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', fmt=".2f",
            xticklabels=columns_to_keep, yticklabels=columns_to_keep, vmin=-1, vmax=1)
```

```python
columns_to_keep = [ 'mean_ElectronAffinity', 'wtd_mean_ElectronAffinity',
       'gmean_ElectronAffinity', 'wtd_gmean_ElectronAffinity',
       'entropy_ElectronAffinity', 'wtd_entropy_ElectronAffinity',
       'range_ElectronAffinity', 'wtd_range_ElectronAffinity',
       'std_ElectronAffinity', 'wtd_std_ElectronAffinity']

corr_matrix = np.corrcoef(data[columns_to_keep], rowvar=False)

# Visualizza la matrice di correlazione utilizzando seaborn
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', fmt=".2f", xticklabels=columns_to_keep, yticklabels=columns_to_keep, vmin=-1, vmax=1)
```

For the next charts, a custom Seaborn theme is needed:

```python
light = {
            "figure.facecolor": "#ffffff",
            "axes.facecolor": "#303030",
            "axes.edgecolor": "#171717",
            "axes.grid": True,
            "grid.color": "#555555",
            "grid.linewidth": 0.5,
            "xtick.color": "#000000",
            "ytick.color": "#000000",
            "axes.labelcolor": "#000000"
            }
        
sns.set_style("dark", rc=light)
```

The following charts are useful to observe the relations between the macro-variables and the critical temperature.

For every variable, the weighted geometric mean is selected, except for the number of elements.

```python
sns.boxplot(
    data=data, x="number_of_elements", y="critical_temp", color="#ff7300",
    flierprops=dict(markerfacecolor='white', markersize=2.5)
)
plt.xlabel("Number of elements")
plt.ylabel("Critical temperature")
plt.title("(1)", color='black')
```
